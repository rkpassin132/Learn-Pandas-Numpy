```python
import pandas as pd

df = pd.read_csv("/content/Mini Project - Marks Adding.csv")
df.head()
```

```python
studentMarks = {}
for row in df.index:
  if(df["Roll Num"][row] in studentMarks): studentMarks[df["Roll Num"][row]] += df["Marks"][row]
  else :  studentMarks[df["Roll Num"][row]] = df["Marks"][row]

# print({"Roll num":list(studentMarks.keys()), "Marks":list(studentMarks.values())})
df2 = pd.DataFrame({"Roll num":list(studentMarks.keys()), "Marks":list(studentMarks.values())})
df2.to_csv("/content/marks.csv")
print(df2.head())
print("File created !")
```
